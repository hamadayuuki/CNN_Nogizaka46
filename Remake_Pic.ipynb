```python
# 2 
# 実行注意
# face フォルダ消えます
# もう一度選別をする必要が出てくる

# Pic ディレクトリから画像を読み込んで顔を切り取ってfaceディレクトリに保存.

import glob   # ファイル内の要素をリストへ格納
import os
import cv2


names = ["白石麻衣","斎藤飛鳥","山下美月","西野七瀬","生田絵梨花"]
out_dir = "./face"
os.makedirs(out_dir, exist_ok=True)   # Python 3.2以降は引数exist_okが追加されており、exist_ok=Trueとすると既に末端ディレクトリが存在している場合もエラーが発生しない。末端ディレクトリが存在していなければ新規作成するし、存在していれば何もしない。

#元画像を取り出して顔部分を正方形で囲み、64×64pにリサイズ、別のファイルにどんどん入れてく
for i in range(len(names)):
    
    # 画像を取り出すファイル名
    in_dir = "./Pic/"+names[i]+"/*.jpg"
    print(in_dir)
    print(type(in_dir))
    
    # Pic フィル内の メンバーファイル の要素を格納
    in_jpg = glob.glob(in_dir)
    
    # 加工後のメンバー別画像 を保存する フォルダを作成 , face フォルダに
    os.makedirs(out_dir +"/"+ names[i], exist_ok=True)
    # print(in_jpg)
    print(len(in_jpg))
    
    # 画像加工
    for num in range(len(in_jpg)):
        
        # 画像読み込み
        image=cv2.imread(str(in_jpg[num]))
        if image is None:
            print("Not open:",num)
            continue

        image_gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # 顔検出のファイル , github からダウンロード必要
        # ダウンロート URL : https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
        cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")   # 顔検出のファイル , github からダウンロード必要
        
        # 顔認識の実行
        # 顔の大きさによって x,y,width,heigh を取得 , list型
        face_list=cascade.detectMultiScale(image_gs, scaleFactor=1.1, minNeighbors=2,minSize=(64,64))
        #print(face_list)
        
        #顔が１つ以上検出された時
        if len(face_list) > 0:
            
            for rect in face_list:
                x,y,width,height=rect
                image = image[rect[1]:rect[1]+rect[3],rect[0]:rect[0]+rect[2]]   # ?
                
                # 画像の辺が 64 以下なら画像を保存しない
                if image.shape[0]<64:
                    continue
                
                image = cv2.resize(image,(64,64))
                
                #保存 , 保存用として作成したファイルに → 保存はファイルがうまく作れていないと画像を保存できない
                fileName=os.path.join(out_dir+"/"+names[i],str(num)+".jpg")
                cv2.imwrite(str(fileName),image)
                print(str(num)+".jpgを保存しました.")
                
        #顔が検出されなかった時
        else:
            print("no face")
            continue
        print(image.shape)
        
    # メンバー別終了通知
    print("\n\n"+names[i]+" : 終了\n\n")
        

# // for i in range(len(names)):


# Next 
# face フォルダの選別
# 下のプログラムでファイルの再作成

```
