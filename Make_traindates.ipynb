```python
# 5

# 訓練データの作成
# 取得し選別した画像を水増しする
# 角度を変える -10,0,10
# その中で 閾値,ぼかし の処理を加える

import os
import cv2
import glob
# n-次元の NumPy 配列を操作する関数を提供します
from scipy import ndimage

names = ["白石麻衣","斎藤飛鳥","山下美月","西野七瀬","生田絵梨花"]
out_dir = "./train"
os.makedirs(out_dir, exist_ok=True)   # Python 3.2以降は引数exist_okが追加されており、exist_ok=Trueとすると既に末端ディレクトリが存在している場合もエラーが発生しない。末端ディレクトリが存在していなければ新規作成するし、存在していれば何もしない。

for name in names:
    in_dir="./face_after/"+name+"/*"
    
    out_dir="./train/"+name
    os.makedirs(out_dir,exist_ok=True)
    in_jpg=glob.glob(in_dir)   # フォルダ内要素全ての ディレクトリ を抜き出す リスト型
    #print(in_jpg[0])   # ./face_after/白石麻衣/77.jpg
    
    img_file_name_list=os.listdir("./face/"+name+"/")   # ?
    
    # in_jpg : 画像データのディレクトリが格納されている
    for i in range(len(in_jpg)):
        
        img=cv2.imread(str(in_jpg[i]))
        
        # 回転
        for ang in [-10,0,10]:
            
            # scipy.ndimage.rotate（入力画像、角度、軸= 1、0、形状変更= True、出力=なし、次数= 3、モード= '定数'、cval = 0.0、プレフィルター= True）
            # 配列を回転させます。
            img_rot=ndimage.rotate(img,ang)
            img_rot=cv2.resize(img_rot,(64,64))
            
            # 回転 : 名前を決めて保存
            filename=os.path.join(out_dir,str(i)+"_"+str(ang)+".jpg")
            cv2.imwrite(str(filename),img_rot)   # 保存
            
            
            
            # 閾値
            
            # cv2.threshold(src, thresh, maxval, type)
            # src : 画像データ 
            # hresh : 閾値 
            # maxval : maxval ( thresh<x<maxval) か 0 ( maxval<x ) かを分ける値 , 
            # typeをcv2.THRESH_TOZEROとした場合は、しきい値より大きい値は元のまま、それ以外の値が0に置き換えられる。
            # 参考 : https://note.nkmk.me/python-numpy-opencv-image-binarization/
            # ? 最後についてる [1]
            img_thr=cv2.threshold(img_rot,100,255,cv2.THRESH_TOZERO)[1]
            
            # 閾値 : 名前を決めて保存
            filename=os.path.join(out_dir,str(i)+"_"+str(ang)+"thr.jpg")
            cv2.imwrite(str(filename),img_thr)
            
            
            # ぼかし
            img_filter=cv2.GaussianBlur(img_rot,(5,5),0)
            filename=os.path.join(out_dir,str(i)+"_"+str(ang)+"filter.jpg")
            
            # ぼかし : 名前を決めて保存
            cv2.imwrite(str(filename),img_filter)
            
            
            
    print(name+" : 終了\n")
            
            
            
            
            
            
            
            
            
            
            
            
            
```
